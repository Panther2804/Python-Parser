```python
#grammar = '<s> = <a>|"b" \n <a> = "aa"<a>|"b"' 
#hier grammatik eintragen
#grammar = '<s> = <minus><zahl>|<zahl> \n <minus> = "-" | \n <zahl> = <ziffer><zahl>|<ziffer>  \n <ziffer> = "1"|"2"|"3"'
grammar = '<s> = "0"<d>"1"<d> \n <d> = "0"<d>"1"|"1"<d>"0"| ' 

print(grammar)
```

```python
#grammatikparser. Zerlegt Grammatik in einzelne Regeln
indef = False #in Regeldefinition ?
i = 0 #momentaner Buchstabe
ins = {} #instructionset
subrule = 0 #string wird an jedem oder geteilt. das sind dann "subrules"
while i < len(grammar): #dieser parser kann sicher umgeschrieben werden, um erweiterte syntak zu akzeptiren. ich nehme psuh requests gerne an.
    print(grammar[i])
    if grammar[i] == " ": #Leerzeichen werden nicht geparst
        i += 1
        print("skipping")
        continue
    print(indef)
    if not indef: #sucht nach nächster variablendeklaration
        print("checking")
        if grammar[i] == "<":
            newins = "" #name der variable
            indef = True
            print("found < @ " + str(i))
            i += 1
            while grammar[i] != ">": #alles bis zum nächsten ">" iste ein variablenname
                #print(grammar[i])
                newins += grammar[i]  
                i += 1
                pass
            print("done with defname")
            while grammar[i] != "=": #überspringt alles bis zur variablendefinition
                i += 1
            i += 1
            ins[newins] = []
            ins[newins].append("") #eröffnet neue subrule
            
            
            
            
    if indef: #in variablendeklaration 
        if grammar[i] == "\n":
            print("done with def")
            subrule = 0
            indef = False
        elif grammar[i] == "|": #neue subrule
            subrule += 1
            ins[newins].append([])
            ins[newins][subrule] = ""
        else:
            ins[newins][subrule] += grammar[i]
        
        
    i += 1
                
print(ins)
```

```python
check = "01111000" #was ist zu überprüfen ? 
global found #das fnkt nicht
found = False

def checker(check,variable,current,level): #überprüfende methode
    empty = False
    rollbackcurrent = current #rollback = dahin züruck, wenn subrule mißt macht
    regel = ins[variable] #ruft anzuwendende regel aus dict auf
    
    levelstring = "" #konstruiert einrückung
    for i in range(0,level):
        levelstring += "|   "
    levelstring += "."
    #try:
    #    print(str(found))
    #except:
    #    print("TF is wrong with python")
        
    print(levelstring + "New Checker check: " + check + " Variable: " + variable + " Current: " + current)
    for i in regel: #einzelne subrules
        print(levelstring + "Currently Working on" + i)
        o = 0
        while o < len(i): #inhalt der subrules
            if i[o] == "<": #wenn andere variable in subrule gefunden
                o += 1
                newvar = ""
                while i[o] != ">":
                    newvar += i[o]
                    o += 1
                print(levelstring + "newvar: " + newvar)
                current += checker(check,newvar,current,level+1) #rufe diese auf
                if "False" in current: #sketchy checker
                    current = rollbackcurrent
                    print(levelstring + " :recursive error - aborting to " + rollbackcurrent)
                    break
                
            
            if i[o] == '"': #wenn string in subrule
                o += 1
                newstring = ""
                while i[o] != '"':
                    newstring += i[o]
                    o += 1
                print(levelstring + "newstring: "+ newstring)
                current += newstring
                
                if current not in check: #wenn angehängter string keinen sinn macht
                    print(levelstring + current + ":Not in check.. aborting... to " + rollbackcurrent)
                    current = rollbackcurrent
                    break
                    
 
                
            o += 1

        if current[len(rollbackcurrent):] == "":
            empty = True 
            print(levelstring + "found empty option" )#+ str(empty))
        elif current[len(rollbackcurrent):] != "" and level != 0: #wenn etwas sinvolles angehängt wurde (an sich müssten hier mehere rückgabewerte für meherere sinnvolle wege akzeptiert werden, viel spaß dabei, das zu implementieren)
            if current in check:
                print(levelstring + "returning with " + current)  
                return current[len(rollbackcurrent):]
        elif current == check and level == 0: #wenn eine richtiger path gefunden wurde
            found = True #hab ich schon erwähnt, dass dieser parser keine leeren rückgabewerte unterstützt ?
            print(levelstring + "FOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUNDFOUND" + str(found))
            return "True"                
        else:
            current = rollbackcurrent #wenn nichts sinvolles gefunden wurde


    if empty == True:
        print(levelstring + "returning empty")
        return ""
    else:
        print(levelstring + "method end")
        return "False"
        
checker(check,"s","",0)    

#grammar = '<s> = |<minus><zahl>|<zahl> \n <minus> = "-" | \n <zahl> = "1"|"2"|"3"' 
```
